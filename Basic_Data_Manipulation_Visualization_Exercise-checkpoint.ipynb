#### A different set of numbers for each location.

```python
def statis(x):
    x = pd.Series(x)
    MinWind = x.min()
    MaxWind = x.max()
    MeanWind = x.mean()
    StdMean = x.std()
    res = [MinWind,MaxWind,MeanWind,StdMean]
    indx = ["Minimum","Maximum","Mean","Standard dev"]
    res = pd.Series(res,index=indx)
    return res
loc_stats = newData.apply(statis)
loc_stats
```
